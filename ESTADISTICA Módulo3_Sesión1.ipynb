# **Variables cualitativas**

Ejercicio: Estado civil

soltero,soltero,soltero,soltero,casado,casado,casado,divorciado

```python
import pandas as pd
```

```python
estados = ["soltero","soltero","divorciado","soltero","soltero","casado","casado","casado"]
```

```python
data = pd.DataFrame(estados)
```

```python
data = data.value_counts() #cuenta las veces que se repite un dato
```

```python
data = data.rename_axis('Estados').to_frame('Cantidad')
```

```python
total = data['Cantidad'].sum()
```

```python
data["Porcentaje"] = data["Cantidad"]*100/total
```

# **Variables cuantitativas Numpy**

Ejercicio:

Edades: 48 40 34 38 48 31 42 25 42 23 34 30


```python
edad = [48,40,34,38,48,31,42,25,42,23,34,30]
```

```python
import numpy as np
```

```python
edad = np.array(edad)
```

```python
#Media: es una medida de tendencia central, es el promedio de los datos.
media = np.mean(edad)
print(media)
```

```python
#Mediana: es una medida de tendencia central, es el centro de los datos.
mediana = np.median(edad)
print(mediana)
```

```python
#Varianza: es una medida de dispersión, cuantifica que tanto estan los dispersos los datos
# en este caso esta medida en años al cuadrado.
varianza = np.var(edad)
print(varianza)
```

```python
#Desviación estandar: es una medida de dispersión
# en este caso esta medida en años.
desviacion = np.std(edad)
print(desviacion)
```
